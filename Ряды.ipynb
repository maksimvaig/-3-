```python
import math

def Coeff(n):            #реккурентное вычисление коэффициентов
    coeff = [-1/4]
    if n == 1:
        return -1/4
    else:
        for i in range(1, n):
            coeff.append(coeff[i-1] * (-1)*((i + 0.5)*i)/(i+1)**2)
        return(coeff[-1])

def Summ(n, x):     #сумма n слагаемых в точке x
    s = 0
    for i in range(1, n+1):
        s += Coeff(i) * x**(-2*i - 1)
    return(s)
    

S = 1/100 * math.asinh(100)
print("Точное значение суммы в первой точке :")
print(S)
print()

print("Наивное суммирование")
n = 1
x = 100
sum_r = math.log(2 * x) / x - Summ(n, x)

while abs(sum_r - S) > 10**(-12):
    n += 1
    sum_r = math.log(2 * x) / x - Summ(n, x)

print("Сумма :", sum_r)
print("Количество слагаемых :", n)
print("Отклонение от точного значения :", abs(S-sum_r))

print()

print("Суммирование в порядке возрастания слагаемых")

terms = []
for i in range(1, n+1):
    terms.append(Coeff(i) * x**(-2*i - 1))

sum_r_1 = math.log(2 * x) / x
for i in range(n):
    sum_r_1 -= terms[terms.index(min(terms))]
    terms[terms.index(min(terms))] = 10**10

print(sum_r_1)

print("Отклонение от точного значения :", abs(S-sum_r_1))

print()

print("Суммирование по Кэхену")

sum_r_2 = math.log(2 * x) / x
comp = 0
y = 0
t = 0
z = 0
i = 0 

while abs(sum_r_2 - S) > 10**(-12):
    i += 1
    y = Coeff(i) * x**(-2*i - 1) - comp
    t = sum_r_2 - y
    z = t - sum_r_2
    comp = z + y
    sum_r_2 = t

print("Сумма :", sum_r_2)
print("Количество слагаемых :", i)
print("Отклонение от точного значения :", abs(S-sum_r_2))

print()

S = 8/9 * math.asinh(9/8)
print("Точное значение суммы во второй точке :")
print(S)
print()

print("Наивное суммирование")
n = 1
x = 9/8
sum_r = math.log(2 * x) / x - Summ(n, x)

while abs(sum_r - S) > 10**(-12):
    n += 1
    sum_r = math.log(2 * x) / x - Summ(n, x)

print("Сумма :", sum_r)
print("Количество слагаемых :", n)
print("Отклонение от точного значения :", abs(S-sum_r))

print()

print("Суммирование в порядке возрастания слагаемых")

terms = []
for i in range(1, n+1):
    terms.append(Coeff(i) * x**(-2*i - 1))

sum_r_1 = math.log(2 * x) / x
for i in range(n):
    sum_r_1 -= terms[terms.index(min(terms))]
    terms[terms.index(min(terms))] = 10**10

print(sum_r_1)

print("Отклонение от точного значения :", abs(S-sum_r_1))

print()

print("Суммирование по Кэхену")

sum_r_2 = math.log(2 * x) / x
comp = 0
y = 0
t = 0
z = 0
i = 0 

while abs(sum_r_2 - S) > 10**(-12):
    i += 1
    y = Coeff(i) * x**(-2*i - 1) - comp
    t = sum_r_2 - y
    z = t - sum_r_2
    comp = z + y
    sum_r_2 = t

print("Сумма :", sum_r_2)
print("Количество слагаемых :", i)
print("Отклонение от точного значения :", abs(S-sum_r_2))
```

```python
import math
import csv
import numpy as np

def SummR(n , x):    #сумма n слагаемых в точке x
    s = 0
    for i in range(n+1):
        s += RecursCoeff(i) / (x + i)
    return s / math.pi

def RecursCoeff(n):       #реккурентное вычисление коэффициентов
    coeff = [math.pi]
    if n == 0:
        return coeff[0]
    if n > 0:
        for i in range(n):
            coeff.append((i + 1 - math.pi)/(i+1) * coeff[i])
    return(coeff[-1])

a = 1
b = 2
step = (b-a)/100
x = a - step
sum_r_spec = 0

tabulation = [[], [], []] 


print("Наивное суммирование :")
for i in range(101):
    x += step
    sum_r_spec = SummR(99, x)
    print(x, ' ', sum_r_spec)
    tabulation[0].append(x)
    tabulation[1].append(sum_r_spec)

print()

print("Суммирование по Кэхену :")

x = a - step

for j in range(101):
    x += step
    sum_r_spec_1 = 0
    comp = 0
    y = 0
    t = 0
    z = 0
    i = -1
    while i <= 99:
        i += 1
        y = RecursCoeff(i) / (math.pi * (x + i)) - comp
        t = sum_r_spec_1 + y
        z = t - sum_r_spec_1
        comp = z - y
        sum_r_spec_1 = t
    print(x, ' ', sum_r_spec_1)
    tabulation[2].append(sum_r_spec_1)

with open("tabulation.csv", 'w') as file:
    writer = csv.writer(file)
    for i in range(101):
        writer.writerow([tabulation[0][i], tabulation[1][i], tabulation[2][i]])
```
